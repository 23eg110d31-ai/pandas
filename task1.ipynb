## TASK-4


```python
#cleaning user_id by replacing empty strings with Nan and converting the column to uppercase
task['user_id']=task['user_id'].replace('',np.nan).str.upper()
task['user_id']
```

## TASK-5

```python
task['age']=pd.to_numeric(task['age'],errors='coerce')
task
```

```python
task.loc[(task['age']<0) | (task['age']>100),'age']=np.nan
task
```
